#### Import required modules/libraries

```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Load Dataset

```python
order_data = pd.read_csv("https://raw.githubusercontent.com/mdaffayusuff/Summer-Course/main/Course-2B/orderdataset%20-%20orderdataset.csv", sep=";")
order_data.head(10)
```

#### Check and Prepare Data

```python
order_data.count()
```

```python
#Check if there's null data
print(order_data.isnull().sum())
```

There are 19 data that have null values in the product_weight_gram column. The data needs to be reviewed to see which products have null values so that proper processing can be done.

```python
# Make a variable that contains only the data with null value
null_weights = order_data[order_data['product_weight_gram'].isnull()]

#Show the data
null_weights[['product_id', 'product_category_name', 'product_weight_gram']]
```

It can be seen that the null data in the product_weight_gram column comes from different product categories. Therefore, the null data will be filled with the mean value of each category.

```python
#count mean for each category that exist in order_data
category_weight_mean = order_data.groupby('product_category_name')['product_weight_gram'].transform('mean')

#fill the null value from order_data
order_data['product_weight_gram'].fillna(category_weight_mean, inplace=True)
```

```python
#check again if there is null data
print(order_data.isnull().sum())
```

It can be seen that for all columns there is no longer any null value data.

```python
# Change purchase_date into date format
order_data['purchase_date'] = pd.to_datetime(order_data['purchase_date'], dayfirst=True)
```

The purchase date column is converted into datetime format so that it can be used to visualize the time series plot

#### Statistical analysis

```python
# Create a 'sales' column by multiplying 'quantity' and 'price'
order_data['sales'] = order_data['quantity'] * order_data['price']

# Calculate top 10 products based on sales
top_products = order_data.groupby('product_id')['sales'].sum().sort_values(ascending=False).head(10)

# Grouped the product_category and sales by product_id
order_data_grouped = order_data.groupby('product_id').agg({
    'product_category_name': 'first', #take the first category, assume that each same product id has the same category
    'sales': 'sum'
}).reset_index()

# Combine top_products with order_data_grouped to get product categories
merged_data = pd.merge(top_products.reset_index(), order_data_grouped, on='product_id', how='left')
merged_data = merged_data[['product_id', 'sales_x', 'product_category_name']]

# Calculate sales statistics
total_sales = order_data['sales'].sum()
average_sales = order_data['sales'].mean()
min_sales = order_data['sales'].min()
max_sales = order_data['sales'].max()

# Print results
print("Top Products:\n", merged_data)
print("\nTotal Sales:", total_sales)
print("Average Sales:", average_sales)
print("Min Sales:", min_sales)
print("Max Sales:", max_sales)
```

*   Product with id 422879e10f46682990de24d770e7f83d which is a gadget category is the product with the highest individual sales.
*   The total sales is 156.053.194.000 rupiah (156 billion rupiah).
*   An average sales amount of 3.121.126,30 rupiah per transaction shows that the average value of each sale is quite high. This can be influenced by the products being sold are relatively high-priced, or that customers are making large purchases.
*   A minimum sales value of 200.000 rupiah and a maximum of 98.940.000 rupiah indicate a large diversity in transaction values.

```python
# Select only numeric columns for correlation calculation
numeric_columns = order_data.select_dtypes(include=[int, float])

# Calculate correlation matrix
correlation_matrix = numeric_columns.corr()

# Print the correlation matrix
print("Correlation Matrix:\n", correlation_matrix)
```

#### Data Visualization

```python
# Visualization Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

Based on the correlation heatmap above, it can be seen that sales and quantity have a fairly strong correlation with 0.69 as well as price and sales which have a correlation value of 0.61.

```python
# Make scatter plot of quantity vs sales
plt.figure(figsize=(10, 6))
sns.scatterplot(x=order_data['quantity'], y=order_data['sales'], hue=order_data['product_category_name'], palette='viridis', s=100)
plt.title('Scatter Plot of Quantity vs Sales')
plt.xlabel('Quantity')
plt.ylabel('Sales')
plt.grid(True)
plt.legend(title='Product Category')
plt.show()
```

Based on the scatter plot above, it can be seen that the correlation between quantity and sales is positive, where the higher the quantity, the higher the sales. In addition, this plot shows the sales behavior, where most products are sold in quantities of less than 10 for each purchase.

```python
# Visualization Top 10 products by Sales
plt.figure(figsize=(10, 6))
sns.barplot(x=top_products.values, y=top_products.index, orient='h')  # orient='h' untuk horizontal
plt.title('Top 10 Products by Sales')
plt.xlabel('Sales')
plt.ylabel('Product ID')
plt.show()
```

Vertical bar charts are used to make it easier to see the ranking of the 10 products with the highest sales. It can be seen that the product with the highest sales is owned by the product with id 422879e10f46682990de24d770e7f83d which is already known previously that this product is in the gadget category.
<br>
Given that the product with the highest sales is in the gadget category, the company can focus on expanding and enhancing the product range within this category. They can conduct market research to identify the latest trends and consumer preferences in the gadget category and then develop products that align with these trends and meet consumer demands.

```python
# Visualization Sales Over Time
plt.figure(figsize=(10, 6))
order_data.set_index('purchase_date')['sales'].plot()
plt.title('Sales Over Time')
plt.xlabel('Purchase Date')
plt.ylabel('Sales')
plt.show()
```

The plot shows the amount sales of each purchase_date. In the plot above, it can be seen that there are days when sales are very high exceeding the average sales. To see the sales trend better, it can be seen in the line chart below

```python
# group data by year and month
order_data['year_month'] = order_data['purchase_date'].dt.to_period('M')

# calculate the monthly quantity
monthly_quantity = order_data.groupby('year_month')['quantity'].sum().reset_index()
monthly_quantity['year_month'] = monthly_quantity['year_month'].dt.to_timestamp()

# calculate the monthly sales
monthly_sales = order_data.groupby('year_month')['sales'].sum().reset_index()
monthly_sales['year_month'] = monthly_sales['year_month'].dt.to_timestamp()

# Making plot side by side horizontaly
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot quantity
axes[0].plot(monthly_quantity['year_month'], monthly_quantity['quantity'], marker='o')
axes[0].set_xlabel('Month-Year')
axes[0].set_ylabel('Total Quantity Sold')
axes[0].set_title('Monthly Sales Quantity Over Time')
axes[0].tick_params(axis='x', rotation=45)

# Plot sales
axes[1].plot(monthly_sales['year_month'], monthly_sales['sales'], marker='o')
axes[1].set_xlabel('Month-Year')
axes[1].set_ylabel('Total Sales')
axes[1].set_title('Monthly Sales Over Time')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()  # Adjust layout to make room for rotated x-axis labels
plt.show()
```

Based on the two line charts above, it can be concluded that the sales trend both based on the quantity of items sold and sales income are both increasing. Sales had a significant increase in November 2017 (2017-11) then experienced a considerable decline in December 2017. This could be due to data that has not been collected at the end of 2017, so it requires further analysis before a decision is made.
<br>
Based on these charts, the company can focus on identifying which product and what makes their sales experienced a significant increase in November 2017. They can also ensure that high-demand products are well-stocked and available, especially during peak sales periods. Other than that, they can implement time-limited promotions in the months on which the sales is lower.


```python
# grouping to see total of quantitiy and sales for each category
order_byCategory = order_data.groupby("product_category_name")
order_byCategory[["quantity", "sales"]].sum()
```

This data is used to identified which category has the most quantity of order

```python
quantity_byCategory = order_byCategory["quantity"].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=quantity_byCategory.values, y=quantity_byCategory.index, orient='h')
plt.title('Category by Quantity Order')
plt.ylabel('Category')
plt.xlabel('Quantity')
plt.show()
```

From the vertical bar chart above, there is no significant difference in the quantity of products sold in each category. The toy category has a slightly higher value than the other categories.

```python
sales_byCategory = order_byCategory["sales"].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=sales_byCategory.values, y=sales_byCategory.index, orient='h')
plt.title('Category by Sales')
plt.ylabel('Category')
plt.xlabel('sales (x10,000,000,000)')
plt.show()
```

Same with the vertical bar chart for category by quantity order, there is no significant difference in the sales of products sold in each category. The toy category has a slightly higher value than the other categories. But the ranking order of categories in quantity and sales has a difference, where based on quantity the rankings are toys, electronics, automotive. While based on sales the ranking is toys, automotive, fashion

```python
# count the frequency of each payment_type
payment_counts = order_data['payment_type'].value_counts()

# Make a pie chart
plt.figure(figsize=(8, 6))
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Payment Types')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

In the pie chart, it can be seen that the distribution of payments is even, with no payment method being particularly popular or particularly less popular.

```python
order_byPayment = order_data.groupby("payment_type")

sales_byPayment = order_byPayment["sales"].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(y=sales_byPayment.index, x=sales_byPayment.values, orient='h')
plt.title('Payment by Sales')
plt.ylabel('Payment')
plt.xlabel('sales')
plt.xlim(3.0e+06, 3.2e+06)
plt.show()
```

However, when viewed based on the payment amount (sales), virtual account have a slightly higher average payment, followed by debit card, credit card, then e-wallet with the least average payment.
